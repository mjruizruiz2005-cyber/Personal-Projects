## Desarrollar Lógica de Recomendación

Diseñar la lógica para el algoritmo recomendador, que tomará un arquetipo de entrada y lo mapeará a composiciones de perfume adecuadas basándose en las estructuras de datos olfativas definidas. Esto podría incluir un sistema de puntuación o reglas de emparejamiento.


### Recommendation Logic Design

To design the recommendation logic, we will use a scoring system that evaluates each olfactory note against an input archetype's preferences and dislikes. The goal is to generate a list of suitable notes or note combinations.

#### 1. Identification using Preferred Criteria:

*   **Preferred_Families**: For each note in `olfactory_df`, we will check if its `Family` is present in the archetype's `Preferred_Families`. A match will contribute positively to the note's score.
*   **Preferred_Characteristics**: Similarly, we will check if a note's `Characteristics` match any of the archetype's `Preferred_Characteristics`. Matches will add to the score.
*   **Gender_Target_Preference**: A note's `Gender_Target` will be compared with the archetype's `Gender_Target_Preference`. Exact matches or 'Unisex' notes for a specific gender preference will receive a positive score. Mismatches will receive a zero or negative score.

#### 2. Exclusion/Penalization using Disliked Notes:

*   **Disliked_Notes**: If a note's `Note` value is found in the archetype's `Disliked_Notes`, it will receive a significant negative penalty, effectively excluding it or making its score very low.

#### 3. Scoring Mechanism:

We will implement a simple additive scoring mechanism. Each criterion met will add a predefined score to the note. The absence of a match will add zero, and a `Disliked_Note` will subtract a large value.

*   **Base Score**: Initialize each note with a score of 0.
*   **Family Match**: +3 points if `Family` is in `Preferred_Families`.
*   **Characteristic Match**: +2 points if `Characteristics` is in `Preferred_Characteristics`.
*   **Gender Match**: +1 point if `Gender_Target` matches `Gender_Target_Preference` or is 'Unisex'.
*   **Disliked Note Penalty**: -10 points if `Note` is in `Disliked_Notes`.

This will result in a ranked list of olfactory notes for the given archetype.

#### 4. Combining and Ranking Notes for Perfume Composition:

Instead of just individual notes, we aim to recommend a 'perfume composition'. This involves selecting a set of notes that complement each other across different `Pyramid_Layer`s (Top, Middle, Base).

1.  **Filter by Score**: First, filter `olfactory_df` to include only notes with a positive score based on the above mechanism.
2.  **Top Notes**: From the filtered set, select the highest-scoring Top Notes.
3.  **Middle Notes**: Select the highest-scoring Middle Notes.
4.  **Base Notes**: Select the highest-scoring Base Notes.
5.  **Composition Generation**: Combine a selection of these high-scoring notes from each layer. The number of notes from each layer can be predefined (e.g., 2 top, 3 middle, 2 base). We can then present these combinations as potential perfume profiles. Further refinement could involve ensuring variety within families or characteristics across layers to create a balanced scent. For a simpler initial approach, we can recommend a list of top-scoring individual notes, categorized by pyramid layer, and suggest that these could form a suitable composition.


### 💡 Design Decision

Now that the recommendation logic is designed, Implement a Python function to calculate scores for each olfactory note based on a given archetype's preferences and dislikes, as outlined in the previous markdown block.


```python
def get_note_scores_for_archetype(archetype_name, olfactory_df, archetype_df):
    # Get the archetype's preferences and dislikes
    archetype_preferences = archetype_df[archetype_df['Archetype_Name'] == archetype_name].iloc[0]

    preferred_families = archetype_preferences['Preferred_Families']
    preferred_characteristics = archetype_preferences['Preferred_Characteristics']
    disliked_notes = archetype_preferences['Disliked_Notes']
    gender_target_preference = archetype_preferences['Gender_Target_Preference']

    # Calculate scores for each olfactory note
    scores = []
    for index, row in olfactory_df.iterrows():
        score = 0

        # 1. Family Match
        if row['Family'] in preferred_families:
            score += 3

        # 2. Characteristic Match
        if row['Characteristics'] in preferred_characteristics:
            score += 2

        # 3. Gender Match
        if row['Gender_Target'] == gender_target_preference or row['Gender_Target'] == 'Unisex':
            score += 1
        # If the gender doesn't match and it's not unisex, penalize lightly
        elif row['Gender_Target'] != 'Unisex' and row['Gender_Target'] != gender_target_preference:
            score -= 0.5 # Small penalty for mismatching gender if not unisex

        # 4. Disliked Note Penalty
        if row['Note'] in disliked_notes:
            score -= 10

        scores.append(score)

    olfactory_df['Score'] = scores
    return olfactory_df.sort_values(by='Score', ascending=False)

print("Function 'get_note_scores_for_archetype' defined successfully.")

# Example usage for 'Modern Minimalist' archetype
scored_olfactory_notes = get_note_scores_for_archetype('Modern Minimalist', olfactory_df.copy(), archetype_df)
print("Example scored olfactory notes for 'Modern Minimalist' archetype (top 5):")
print(scored_olfactory_notes.head())
```

### 💡 Design Decision

Now that individual notes can be scored, Implement a function to generate perfume compositions by selecting the top-scoring notes from each pyramid layer (Top, Middle, Base) for a given archetype, as outlined in the recommendation logic design.


```python
def recommend_perfume_composition(archetype_name, olfactory_df, archetype_df, top_n_per_layer=2):
    # Get scored olfactory notes for the archetype
    scored_notes = get_note_scores_for_archetype(archetype_name, olfactory_df.copy(), archetype_df)

    # Filter for notes with a positive score
    positive_score_notes = scored_notes[scored_notes['Score'] > 0]

    if positive_score_notes.empty:
        return "No suitable notes found for this archetype with a positive score."

    recommended_composition = {
        'Top Notes': [],
        'Middle Notes': [],
        'Base Notes': []
    }

    # Select top notes for each pyramid layer
    for layer in ['Top', 'Middle', 'Base']:
        layer_notes = positive_score_notes[positive_score_notes['Pyramid_Layer'] == layer]
        # Sort by score and get top_n_per_layer
        top_notes = layer_notes.head(top_n_per_layer)
        recommended_composition[f'{layer} Notes'] = top_notes[['Note', 'Score']].to_dict(orient='records')

    return recommended_composition

print("Function 'recommend_perfume_composition' defined successfully.")

# Example usage for 'Modern Minimalist' archetype
composition_modern_minimalist = recommend_perfume_composition('Modern Minimalist', olfactory_df.copy(), archetype_df)
print("\nRecommended perfume composition for 'Modern Minimalist':")
print(composition_modern_minimalist)

# Example usage for 'Classic Romantic' archetype
composition_classic_romantic = recommend_perfume_composition('Classic Romantic', olfactory_df.copy(), archetype_df)
print("\nRecommended perfume composition for 'Classic Romantic':")
print(composition_classic_romantic)
```

#### 1. Function `calculate_harmony_score(composition, compatibility_matrix_df, olfactory_df)`:

This function will take a perfume composition (a dictionary of notes by pyramid layer), the `compatibility_matrix_df`, and the `olfactory_df` as input.

*   **Iterate through note pairs**: For each possible pair of notes within the composition, or specifically focusing on inter-layer compatibility (e.g., Top-Middle, Middle-Base, Top-Base), we will find their respective olfactory families.
*   **Retrieve compatibility score**: Look up the compatibility score between these two families in the `compatibility_matrix_df`.
    *   To handle asymmetric pairs, ensure that `compatibility_matrix_df` includes both `(Family1, Family2)` and `(Family2, Family1)` pairs, or implement a lookup that checks both.
*   **Aggregate scores**: Sum or average these compatibility scores to get an overall harmony score for the entire composition.

#### 2. Integration into `recommend_perfume_composition`:

*   After generating potential `top_n_per_layer` notes for each pyramid layer, instead of directly returning them, we will generate multiple candidate compositions.
*   For each candidate composition, we will calculate its `harmony_score` using the new function.
*   The final recommended composition will be the one that has the highest combined score (initial individual note scores + harmony score) or just the highest harmony score among the highly-rated individual notes.

#### Scoring Mechanism for Harmony:

*   For simplicity, we can calculate an average compatibility score for all unique pairs of notes across different pyramid layers in a composition.
*   For example, if a composition has notes (N1, N2, N3) from Top, Middle, Base layers respectively:
    *   Calculate compatibility between (Family(N1), Family(N2))
    *   Calculate compatibility between (Family(N2), Family(N3))
    *   Calculate compatibility between (Family(N1), Family(N3))
*   The harmony score could be the average of these compatibility scores.
*   This harmony score will then be used to refine the selection of the best perfume compositions.


### 💡 Design Decision

Now, Implement the `calculate_harmony_score` function based on the design outlined in the previous markdown block. This function will calculate an overall harmony score for a given perfume composition by averaging the compatibility scores between all unique pairs of olfactory families present in the composition.


```python
def calculate_harmony_score(composition, compatibility_matrix_df, olfactory_df):
    all_notes_in_composition = []
    for layer in composition:
        for note_info in composition[layer]:
            all_notes_in_composition.append(note_info['Note'])

    # Get the families for all notes in the composition
    families_in_composition = []
    for note in all_notes_in_composition:
        # Ensure the note exists in olfactory_df and handle potential multiple entries if needed (take the first)
        family_row = olfactory_df[olfactory_df['Note'] == note]
        if not family_row.empty:
            families_in_composition.append(family_row['Family'].iloc[0])

    # Get unique families to avoid redundant calculations for notes of the same family
    unique_families = list(set(families_in_composition))

    if len(unique_families) < 2:
        return 0.0 # Cannot calculate harmony with less than two unique families

    harmony_scores = []
    # Calculate compatibility for all unique pairs of families
    for i in range(len(unique_families)):
        for j in range(i + 1, len(unique_families)):
            family1 = unique_families[i]
            family2 = unique_families[j]

            # Look up compatibility score, checking both (family1, family2) and (family2, family1)
            score_row = compatibility_matrix_df[
                ((compatibility_matrix_df['Family1'] == family1) & (compatibility_matrix_df['Family2'] == family2)) |
                ((compatibility_matrix_df['Family1'] == family2) & (compatibility_matrix_df['Family2'] == family1))
            ]

            if not score_row.empty:
                harmony_scores.append(score_row['Compatibility_Score'].iloc[0])
            else:
                # Default low compatibility if no specific score is defined
                harmony_scores.append(0.2) # Default score for unlisted pairs

    if not harmony_scores:
        return 0.0 # No harmony scores calculated

    return sum(harmony_scores) / len(harmony_scores)

print("Function 'calculate_harmony_score' defined successfully.")

# Example usage with an existing composition (e.g., composition_modern_minimalist)
example_composition = composition_modern_minimalist
harmony = calculate_harmony_score(example_composition, compatibility_matrix_df, olfactory_df)
print(f"Harmony score for Modern Minimalist composition: {harmony}")
```

### Modificación de `recommend_perfume_composition` para Integrar Armonía

Para integrar la puntuación de armonía en el algoritmo de recomendación, modificaremos la función `recommend_perfume_composition` de la siguiente manera:

1.  **Generación de Candidatos**: En lugar de seleccionar directamente `top_n_per_layer` notas y formar una única composición, generaremos múltiples combinaciones de notas para cada capa (Top, Middle, Base) a partir de las notas con puntuación positiva.
    *   Para simplificar y mantener un número manejable de combinaciones, seleccionaremos las `top_k` notas con mayor puntuación para cada capa (donde `k` podría ser 2 o 3, por ejemplo). Luego, generaremos todas las combinaciones posibles tomando una nota de Top, una de Middle y una de Base.

3.  **Puntuación Final de Composición**: Combinaremos la puntuación individual de las notas y la puntuación de armonía para obtener una `final_composition_score`.
    *   Una forma de combinar podría ser: `final_composition_score = (peso_individual * individual_note_score) + (peso_armonia * harmony_score)`.
    *   Para esta implementación inicial, podemos simplemente sumar ambas, o dar un peso mayor a la armonía para enfatizarla.

4.  **Selección de la Mejor Composición**: La función devolverá la composición candidata que tenga la `final_composition_score` más alta.

### Consideraciones para la Generación de Candidatos:

*   **Flexibilidad en la selección**: Asegurarse de que si una capa tiene menos de `top_k` notas con puntuación positiva, se utilicen todas las disponibles. El `top_n_per_layer` original ya maneja esto hasta cierto punto.


### 💡 Design Decision

I will modify the `recommend_perfume_composition` function to incorporate the harmony scoring logic. This will involve generating candidate compositions from the top-scoring notes for each layer, calculating an overall score for each candidate (combining individual note scores and harmony scores), and then selecting the best composition.


```python
import itertools # Import itertools for generating combinations

def recommend_perfume_composition_with_harmony(archetype_name, olfactory_df, archetype_df, compatibility_matrix_df, top_k_for_combinations=3):
    # Get scored olfactory notes for the archetype
    scored_notes = get_note_scores_for_archetype(archetype_name, olfactory_df.copy(), archetype_df)

    # Filter for notes with a positive score
    positive_score_notes = scored_notes[scored_notes['Score'] > 0]

    if positive_score_notes.empty:
        return "No suitable notes found for this archetype with a positive score."

    # Get top_k notes for each pyramid layer to generate combinations
    top_notes_candidates = positive_score_notes[positive_score_notes['Pyramid_Layer'] == 'Top'].head(top_k_for_combinations)
    middle_notes_candidates = positive_score_notes[positive_score_notes['Pyramid_Layer'] == 'Middle'].head(top_k_for_combinations)
    base_notes_candidates = positive_score_notes[positive_score_notes['Pyramid_Layer'] == 'Base'].head(top_k_for_combinations)

    # If any layer is empty, return an appropriate message or handle it gracefully
    if top_notes_candidates.empty or middle_notes_candidates.empty or base_notes_candidates.empty:
        return f"Could not find enough suitable notes for all layers for archetype '{archetype_name}'."

    # Generate all possible combinations of compositions
    all_candidate_compositions = []
    for top_note_row in top_notes_candidates.itertuples():
        for middle_note_row in middle_notes_candidates.itertuples():
            for base_note_row in base_notes_candidates.itertuples():
                candidate_composition = {
                    'Top Notes': [{'Note': top_note_row.Note, 'Score': top_note_row.Score}],
                    'Middle Notes': [{'Note': middle_note_row.Note, 'Score': middle_note_row.Score}],
                    'Base Notes': [{'Note': base_note_row.Note, 'Score': base_note_row.Score}]
                }
                all_candidate_compositions.append(candidate_composition)

    if not all_candidate_compositions:
        return "No valid combinations could be generated."

    best_composition = None
    max_final_score = -1

    # Evaluate each candidate composition
    for comp in all_candidate_compositions:
        # Calculate individual note score sum
        individual_note_score_sum = 0
        for layer_notes in comp.values():
            for note_info in layer_notes:
                individual_note_score_sum += note_info['Score']

        # Calculate harmony score
        harmony_score = calculate_harmony_score(comp, compatibility_matrix_df, olfactory_df)

        # Combine scores. We can weight them if necessary. For now, simple sum + a factor for harmony
        # Giving harmony a significant weight as it's a new, important factor
        final_composition_score = individual_note_score_sum + (harmony_score * 10) # Multiply harmony by 10 to give it more impact

        if final_composition_score > max_final_score:
            max_final_score = final_composition_score
            best_composition = comp

    # Add the harmony score to the best composition for display
    if best_composition:
        best_composition['Harmony_Score'] = calculate_harmony_score(best_composition, compatibility_matrix_df, olfactory_df)
        best_composition['Final_Total_Score'] = max_final_score

    return best_composition

print("Function 'recommend_perfume_composition_with_harmony' defined successfully.")

# Example usage for 'Modern Minimalist' archetype with harmony
composition_modern_minimalist_harmony = recommend_perfume_composition_with_harmony('Modern Minimalist', olfactory_df.copy(), archetype_df, compatibility_matrix_df)
print("\nRecommended perfume composition for 'Modern Minimalist' (with harmony scoring):")
print(composition_modern_minimalist_harmony)

# Example usage for 'Classic Romantic' archetype with harmony
composition_classic_romantic_harmony = recommend_perfume_composition_with_harmony('Classic Romantic', olfactory_df.copy(), archetype_df, compatibility_matrix_df)
print("\nRecommended perfume composition for 'Classic Romantic' (with harmony scoring):")
print(composition_classic_romantic_harmony)
```

## Implementar Métrica de Similitud Fuzzy

Desarrollar una función para calcular una 'puntuación de similitud' o 'afinidad' entre las preferencias de un nuevo usuario y cada arquetipo existente en `archetype_df`, utilizando la superposición de familias preferidas, características y penalizando las notas no deseadas.

#### Recommendation Logic Design - Fuzzy Similarity Metric

To implement the fuzzy similarity metric, we will create a function `calculate_fuzzy_similarity(new_user_prefs, archetype_profile)` that compares the new user's preferences with a given archetype's profile and returns a similarity score.

#### 1. Overlap Score for Preferred Families and Characteristics:

*   **Preferred_Families**: Calculate the proportion of the new user's `Preferred_Families` that are also present in the `archetype_profile['Preferred_Families']`. This will be a score between 0 and 1.
    *   `overlap_families = len(set(new_user_prefs['Preferred_Families']) & set(archetype_profile['Preferred_Families'])) / len(new_user_prefs['Preferred_Families'])`
*   **Preferred_Characteristics**: Similarly, calculate the overlap for `Preferred_Characteristics`.
    *   `overlap_characteristics = len(set(new_user_prefs['Preferred_Characteristics']) & set(archetype_profile['Preferred_Characteristics'])) / len(new_user_prefs['Preferred_Characteristics'])`

#### 2. Penalty for Disliked Notes:

*   **Disliked_Notes**: If any of the new user's `Disliked_Notes` are present in the `archetype_profile['Preferred_Families']` or `archetype_profile['Preferred_Characteristics']` (we'll need a way to map notes to families/characteristics, or directly check if archetype's preferred notes are disliked by user), a penalty will be applied.
    *   For simplicity, we can check if any `archetype_profile['Preferred_Families']` are known to be associated with user's `Disliked_Notes`' families or if archetype's `Preferred_Characteristics` align with notes user dislikes. A simpler approach for now is to check if any of the archetype's *key features* (families, characteristics) are explicitly associated with user's disliked notes.
    *   A more direct approach is to check if any notes typically associated with the archetype's `Preferred_Families` or `Preferred_Characteristics` are in the user's `Disliked_Notes` list. For the initial implementation, we will directly check if notes from `archetype_profile['Preferred_Families']` are associated with `Disliked_Notes` of the new user. Or, more simply, if any *actual notes* that would be recommended by an archetype are in the user's `Disliked_Notes`.
    *   *Refinement for Disliked Notes*: We can penalize if any of the *archetype's preferred notes* (from `archetype_profile['Preferred_Families']` or `archetype_profile['Preferred_Characteristics']`) are directly mentioned in `new_user_prefs['Disliked_Notes']` in `olfactory_df` (by looking up note families/characteristics).
    *   For a first pass, let's simplify: A penalty will be applied for each of the new user's `Disliked_Notes` that are part of the `olfactory_df` and whose 'Family' or 'Characteristics' are among the `archetype_profile`'s `Preferred_Families` or `Preferred_Characteristics` respectively. This is getting complex. Let's simplify further: a penalty if a `Disliked_Note` by the user is a prominent note in the archetype's suggested profile.
    *   **Simpler Disliked Notes Penalty**: Iterate through the `new_user_prefs['Disliked_Notes']`. For each disliked note, check its family and characteristics from `olfactory_df`. If this note's family is in the archetype's `Preferred_Families` OR its characteristics are in the archetype's `Preferred_Characteristics`, apply a penalty.

#### 3. Gender Preference Match:

*   **Gender_Target_Preference**: A score will be given if `new_user_prefs['Gender_Target_Preference']` matches `archetype_profile['Gender_Target_Preference']` or if either is 'Unisex'.

#### 4. Combining Scores:

*   The final similarity score will be a weighted sum of the overlap scores, minus any penalties.
    *   `similarity_score = (weight_families * overlap_families) + (weight_characteristics * overlap_characteristics) + (weight_gender * gender_match_score) - (weight_disliked * total_penalty_from_disliked_notes)`

#### Weights (initial proposal):
*   `weight_families = 0.4`
*   `weight_characteristics = 0.3`
*   `weight_gender = 0.2`
*   `penalty_per_disliked_overlap = 0.2`


### 💡 Design Decision

Now that the design for the fuzzy similarity metric is outlined, Implement the `calculate_fuzzy_similarity` function as described, comparing a new user's preferences against an archetype's profile based on preferred families, characteristics, disliked notes, and gender preferences.


```python
def calculate_fuzzy_similarity(new_user_prefs, archetype_profile, olfactory_df):
    similarity_score = 0

    # Weights for combining scores
    weight_families = 0.4
    weight_characteristics = 0.3
    weight_gender = 0.2
    penalty_per_disliked_overlap = 0.2

    # 1. Overlap Score for Preferred Families
    new_user_families = set(new_user_prefs['Preferred_Families'])
    archetype_families = set(archetype_profile['Preferred_Families'])
    overlap_families = len(new_user_families.intersection(archetype_families)) / len(new_user_families) if new_user_families else 0
    similarity_score += weight_families * overlap_families

    # 2. Overlap Score for Preferred Characteristics
    new_user_characteristics = set(new_user_prefs['Preferred_Characteristics'])
    archetype_characteristics = set(archetype_profile['Preferred_Characteristics'])
    overlap_characteristics = len(new_user_characteristics.intersection(archetype_characteristics)) / len(new_user_characteristics) if new_user_characteristics else 0
    similarity_score += weight_characteristics * overlap_characteristics

    # 3. Gender Preference Match
    gender_match_score = 0
    if new_user_prefs['Gender_Target_Preference'] == archetype_profile['Gender_Target_Preference']:
        gender_match_score = 1
    elif archetype_profile['Gender_Target_Preference'] == 'Unisex' or new_user_prefs['Gender_Target_Preference'] == 'Unisex':
        # If either is unisex, it's considered a partial match
        gender_match_score = 0.5
    similarity_score += weight_gender * gender_match_score

    # 4. Penalty for Disliked Notes
    disliked_penalty = 0
    for disliked_note_name in new_user_prefs['Disliked_Notes']:
        # Find the family and characteristics of the disliked note
        note_info = olfactory_df[olfactory_df['Note'] == disliked_note_name]
        if not note_info.empty:
            disliked_note_family = note_info['Family'].iloc[0]
            disliked_note_characteristic = note_info['Characteristics'].iloc[0]

            # Check if archetype's preferred families or characteristics overlap with the disliked note's attributes
            if disliked_note_family in archetype_families or disliked_note_characteristic in archetype_characteristics:
                disliked_penalty += penalty_per_disliked_overlap

    similarity_score -= disliked_penalty

    return max(0, similarity_score) # Ensure score doesn't go below zero

print("Function 'calculate_fuzzy_similarity' defined successfully.")

# Example usage for 'Modern Minimalist' archetype
modern_minimalist_profile = archetype_df[archetype_df['Archetype_Name'] == 'Modern Minimalist'].iloc[0]
fuzzy_similarity_modern_minimalist = calculate_fuzzy_similarity(new_user_preferences, modern_minimalist_profile, olfactory_df)
print(f"Fuzzy similarity for 'Modern Minimalist' archetype: {fuzzy_similarity_modern_minimalist}")
```

## Desarrollar una Función de Identificación de Arquetipos Fuzzy

Crear una función principal que tome las preferencias de un nuevo usuario, utilice la métrica de similitud para puntuar la afinidad con todos los arquetipos existentes, y devuelva los N arquetipos más similares.

#### Recommendation Logic Design - Fuzzy Archetype Identification

We will create a function `identify_fuzzy_archetype(new_user_prefs, archetype_df, olfactory_df, top_n=1)` that takes the new user's preferences, the DataFrame of existing archetypes, the `olfactory_df` (for disliked note penalty calculation), and returns the top N most similar archetypes.

#### 1. Iterate Through Archetypes:

*   The function will loop through each archetype in the `archetype_df`.
*   For each archetype, it will extract its profile (e.g., `Preferred_Families`, `Preferred_Characteristics`, `Disliked_Notes`, `Gender_Target_Preference`).

#### 2. Calculate Similarity Score:

*   For each archetype profile, it will call the `calculate_fuzzy_similarity` function (which was just implemented) to compute a similarity score between the `new_user_prefs` and the current archetype's profile.

#### 3. Store Scores:

*   The similarity score for each archetype will be stored, along with the archetype's name, in a list of tuples or dictionaries.

#### 4. Rank and Select Top N:

*   Once all archetypes have been scored, the list will be sorted in descending order based on their similarity scores.
*   The function will then return the top `top_n` archetypes from this sorted list. If `top_n` is 1, it will return the single most similar archetype.

#### Output Structure:

The function should return a list of dictionaries, where each dictionary contains the 'Archetype_Name' and its 'Similarity_Score'.


### 💡 Design Decision

Now that the design for the fuzzy archetype identification function is outlined, Implement the `identify_fuzzy_archetype` function to score all existing archetypes against the new user's preferences and return the top N most similar archetypes.


```python
def identify_fuzzy_archetype(new_user_prefs, archetype_df, olfactory_df, top_n=1):
    archetype_similarities = []

    for index, archetype_profile in archetype_df.iterrows():
        similarity_score = calculate_fuzzy_similarity(new_user_prefs, archetype_profile, olfactory_df)
        archetype_similarities.append({
            'Archetype_Name': archetype_profile['Archetype_Name'],
            'Similarity_Score': similarity_score
        })

    # Sort archetypes by similarity score in descending order
    sorted_archetypes = sorted(archetype_similarities, key=lambda x: x['Similarity_Score'], reverse=True)

    return sorted_archetypes[:top_n]

print("Function 'identify_fuzzy_archetype' defined successfully.")

# Example usage to find the top 1 most similar archetype
top_archetype = identify_fuzzy_archetype(new_user_preferences, archetype_df.copy(), olfactory_df)
print("\nTop 1 most similar archetype for new user preferences:")
print(top_archetype)
```

```python
def recommend_perfume_for_new_user(new_user_prefs, archetype_df, olfactory_df, compatibility_matrix_df, top_k_for_combinations=3):
    # 1. Identify the most similar archetype for the new user
    identified_archetypes = identify_fuzzy_archetype(new_user_prefs, archetype_df, olfactory_df, top_n=1)

    if not identified_archetypes:
        return "No suitable archetype found for these preferences."

    # Extract the name of the most similar archetype
    best_archetype_name = identified_archetypes[0]['Archetype_Name']
    similarity_score = identified_archetypes[0]['Similarity_Score']

    print(f"Identified best archetype for new user: '{best_archetype_name}' with similarity score: {similarity_score:.2f}")

    # 2. Generate the harmonized perfume composition using the identified archetype
    # We call the existing function with the identified archetype's name
    recommendation = recommend_perfume_composition_with_harmony(
        best_archetype_name,
        olfactory_df.copy(),
        archetype_df,
        compatibility_matrix_df,
        top_k_for_combinations
    )

    return {
        'Identified_Archetype': best_archetype_name,
        'Archetype_Similarity_Score': similarity_score,
        'Recommended_Composition': recommendation
    }

print("Function 'recommend_perfume_for_new_user' defined successfully.")
```
